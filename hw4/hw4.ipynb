```bash
git clone https://github.com/jonathanhar/diminutives.db 2>/dev/null || true
cp -f diminutives.db/female_diminutives.csv diminutives.db/male_diminutives.csv .
ls -1 female_diminutives.csv male_diminutives.csv
```

```bash
awk -F, '{print $1}' female_diminutives.csv male_diminutives.csv | sort
```

```bash
awk -F, '{print $1, NF-1}' female_diminutives.csv male_diminutives.csv
```

```bash
awk -F, '{print $1, NF-1}' female_diminutives.csv male_diminutives.csv | sort -k2,2nr | head -5
```

```bash
awk -F, '{c[toupper(substr($1,1,1))]++} END {for(i=65;i<=90;i++){ch=sprintf("%c", i); printf "%s: %d\n", ch, c[ch]+0}}' female_diminutives.csv male_diminutives.csv
```

```bash
awk -F, '{for(i=2;i<=NF;i++) if(length($i)==3) print $1, $i}' female_diminutives.csv male_diminutives.csv
```

```bash
awk -F, '{sum+=NF-1; n++} END {if(n) print sum/n}' female_diminutives.csv male_diminutives.csv
```

```bash
awk -F, '{sum+=length($1); n++} END {if(n) print sum/n}' female_diminutives.csv male_diminutives.csv
```

```bash
awk -F, '$1 ~ /^A/ {print $1}' female_diminutives.csv male_diminutives.csv
```

```bash
awk -F, '{if($1 ~ /^A/) print $1; for(i=2;i<=NF;i++) if($i ~ /^A/) print $1, $i}' female_diminutives.csv male_diminutives.csv
```

```bash
awk -F, '{total+=NF} END {print total}' female_diminutives.csv male_diminutives.csv
```
